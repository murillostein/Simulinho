```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# Importando os dados

```python
gabarito = pd.read_excel("Gabarito\gabarito_teste.xlsx")
# veio com colunas à mais, a linha a seguir exclui tais colunas
gabarito = gabarito.drop(gabarito.columns[[5]], axis = 1)

gabarito
```

```python
# respostas das questões objetivas
respostas_alunos = pd.read_excel("Gabarito/respostas_alunos.xlsx")
respostas_alunos
```

```python
# dados pessoais dos alunos
dados_pessoais = pd.read_excel("Gabarito\dados_alunos.xlsx")
# veio com colunas à mais, a linha a seguir exclui tais colunas
dados_pessoais = dados_pessoais.drop(dados_pessoais.columns[[5, 6, 7,8 ]], axis = 1)

dados_pessoais.to_pickle("dados-relatorio-alunos/dados_pessoais.pkl")

dados_pessoais
```

```python
# quebra as respostas para se ter uma resposta por pessoa por linha:
# pd.melt faz exatamente isso. as id_vars são as variáveis de identificação, que não devem ser quebradas por linha
data = pd.melt(respostas_alunos, id_vars=['nome'], var_name='questao', value_name='resposta')
data
```

```python
# precisamos de uma coluna dizendo qual é a lingua estrangeira escolhida pelo aluno
# embora no simulinho 2021 não existam qeustões de linguas
data = pd.merge(data, dados_pessoais, on='nome')
data = data.drop(columns=['cpf', 'curso', 'paa'], axis=1)
data.head(10)
```

```python
# precisamos juntar as respostas de cada aluno ao gabarito
# merge faz justamente isso
data = pd.merge(data, gabarito, on='questao')
data
```

```python
data = data.rename(columns={'Materias':'materia'})
```

```python
# a correção (acerto ou erro) nada mais é que a comparação entre a resposta e o gabarito
data['correcao'] = data['resposta'] == data['gabarito']


data
```

```python
# nome -> materia -> questao -> resposta -> gabarito -> dificuldade -> assunto


data_clean = data[['nome','materia', 'questao','resposta','gabarito','assunto','dificuldade']].sort_values(['nome', 'questao']).set_index(['nome','materia','questao'])


# salvando como pickle para o relatorio
data_clean.to_pickle('dados-relatorio-alunos/data.pkl')

#df_questoes_alunos = df_questions[['Nome','Matéria','Questão','Resposta','Alternativa correta','Tema']].sort_values(['Nome', 'Questão']).set_index(['Nome','Matéria','Questão'])

data_clean 
```

## Alunos

```python
# calcula o total de questões certas por máteria, quantas questões tinham por matéria, e média por matéria

# agg permite agregar multiplas funções ao groupby
acertos_aluno = data.groupby(['nome','materia'])['correcao'].agg(['sum','count','mean'])
acertos_aluno = acertos_aluno.rename(columns={"sum": "Total Acertos", "mean": "Média Individual", 'count': "Total questoes"})
```

```python
# comparar a média de acerto das questões dos alunos com a turma geral

# media geral do einstein
media_acerto_materia = data.groupby('materia').mean()

media_acerto_materia = media_acerto_materia.rename(columns={"correcao": "Media Geral"})

# adicionando a media geral aos dados individuais
acertos_aluno = acertos_aluno.join(media_acerto_materia, on='materia')
acertos_aluno
```

```python
acertos_aluno = acertos_aluno.unstack()
acertos_aluno.head()
```

```python
acertos_aluno[('Total Acertos', 'Total')] = acertos_aluno['Total Acertos'].T.sum()
```

```python
acertos_aluno[('Total questoes', 'Total')] = 50
```

```python
acertos_aluno[('Média Individual', 'Total')] = acertos_aluno[('Total Acertos', 'Total')] / 50
```

```python
acertos_aluno
```

```python
acertos_aluno = acertos_aluno.stack()
media_acertos_obj = data['correcao'].mean()
acertos_aluno = acertos_aluno.fillna(value=media_acertos_obj)
acertos_aluno
```

```python
# manipulando a coluna de total de acertos pra mostrar no formato -> acertos/total de questoes
acertos_aluno[['Total Acertos', 'Total questoes']] = acertos_aluno[['Total Acertos', 'Total questoes']].astype(str)

acertos_aluno['Total de Acertos'] = acertos_aluno[['Total Acertos', 'Total questoes']].agg('/'.join, axis=1)

acertos_aluno = acertos_aluno.drop(columns =['Total Acertos', 'Total questoes'])
```

```python
# reordenando as colunas
acertos_aluno = acertos_aluno[['Total de Acertos',  "Média Individual", "Media Geral"]]
```

```python
# salvando como pickle para o relatorio
acertos_aluno.to_pickle('dados-relatorio-alunos/acertos_aluno.pkl')

acertos_aluno
```

```python
# media de acerto do aluno por dificuldade de questao e por materia
media_por_dificuldade = data.groupby(['nome','materia','dificuldade'])['correcao'].agg(['sum','count','mean'])
media_por_dificuldade = media_por_dificuldade.rename(columns={"sum": "Total Acertos", "mean": "Média Individual", 'count': "Total questoes"})

# media geral do einstein
media_geral_dificuldade = data.groupby(['materia','dificuldade']).mean()


# adicionando a media geral aos dados individuais
media_por_dificuldade = media_por_dificuldade.join(media_geral_dificuldade, on=['materia','dificuldade'],how='left', lsuffix='_left', rsuffix='_right')
media_por_dificuldade = media_por_dificuldade.rename(columns={'correcao_left': 'Média Individual', 'correcao_right':'Media Geral'})

media_por_dificuldade = media_por_dificuldade.rename(columns={'correcao': 'Média Geral'})

media_por_dificuldade
```

```python
# media geral do einstein por dificuldade das questoes para os alunos compararem
media_por_dificuldade = data.groupby(['materia','dificuldade']).mean()

media_por_dificuldade
```

```python
# agrupar as questoes por materia e por tema

# podem existir matérias que não tem os tres tipos(fácil, médio, dificil)
acertos_por_assunto = data.groupby(['nome','materia', 'assunto'])['correcao'].agg(['sum','count','mean'])

acertos_por_assunto = acertos_por_assunto.rename(columns={"sum": "Total Acertos", "mean": "Média Individual", 'count': "Total questoes"})

# salvando como pickle para o relatorio
acertos_por_assunto.to_pickle('dados-relatorio-alunos/acertos_por_assunto.pkl')

acertos_por_assunto
```

## Pontuação Total

```python
# juntando a pontuação das questões objetivas com a nota da redação

# pegando a soma de acertos de cada aluno
acertos_aluno_total = data.groupby(['nome']).sum()

# juntando os dados das notas de redacao com os dados das questoes objetivas
pontuacao_total = dados_redacao
pontuacao_total = pontuacao_total.join(acertos_aluno_total, on='nome')

# para a pontuação total, não precisamos das colunas de critérios e comentarios
pontuacao_total = pontuacao_total.drop(labels=['Norma culta','Adequação ao tema','Argumentação','Coesão', 'Proposta de intervenção'
,'comentarios', 'comentarios.1','comentarios.2','comentarios.3','comentarios.4', 'comentarios.5'], axis =1)

# aqui considera que cada questao vale 10 pontos
pontuacao_total['Nota total'] = pontuacao_total['Nota'] + pontuacao_total['correcao']*10

# salvando como pickle para o relatorio
pontuacao_total.to_pickle('dados-relatorio-alunos/pontuacao_total.pkl')

pontuacao_total
```

```python
# calcular a % de acerto de cada aluno
# como cada questão vale 10, e tem 50 questoes, logo o peso da prova objetiva é 500
# redação vale 1000
# nota máxima é 1500
pontuacao_total['% de acerto'] = pontuacao_total['Nota total'] / 1500

# salvando como pickle para o relatorio
pontuacao_total.to_pickle('dados-relatorio-alunos/pontuacao_total.pkl')

pontuacao_total
```

```python
# ver a colocação de cada aluno

colocacao = pontuacao_total.sort_values(by='Nota total', ascending=False)

# como os indices ficam fora de ordem depois de colocar as notas em valor decrescente, é necessário resetar o valor do indices
# de 'colocacao'
colocacao = colocacao.reset_index(drop=True)

# salvando como pickle para o relatorio
colocacao.to_pickle('dados-relatorio-alunos/colocacao.pkl')

colocacao
```

# Gráficos e relatórios

```python
# gráfico para a media de acerto de cada questão, separado por matéria

# media de acertos em cada questão
media_acerto_questao = data.groupby(['materia', 'questao']).mean()

# como 'media_acerto_questao' tem indice igual ao nome das materias, uma melhor prática seria criar uma nova coluna especifica
# para o indice
media_acerto_questao = media_acerto_questao.reset_index()
# lista de materias do simulinho
materias = ['Biologia','Física','Geografia','História','Matemática','Português','Química']

# o loop passa por cada materia do vetor acima, comparar se o valor é igual à materia do conjunto 'media_Acerto_questao'
# e plota o gráfico com a média de acertos na referida materia
for materia in materias:
    correcao = media_acerto_questao[media_acerto_questao['materia'] == materia]
    plt.figure()
    plt.bar(correcao['questao'], correcao['correcao'])
    plt.title(materia)

plt.show()
```

```python
# gráfico da média de acerto de todas as matérias

media_acerto_materia = data.groupby(['materia']).mean()

# como 'media_acerto_questao' tem indice igual ao nome das materias, uma melhor prática seria criar uma nova coluna especifica
# para o indice
media_acerto_materia = media_acerto_materia.reset_index()

sns.barplot(x=media_acerto_materia['materia'], y=media_acerto_materia['correcao'])
plt.xticks(rotation=45)
plt.show()
```
